```python
import os
import json
import re
from bs4 import BeautifulSoup
import requests
import unicodedata

# Specify the directory path to save the JSON files
directory = '/content/drive/MyDrive/Bigdata_Project/TongHopBaoCaoDoanhNghiepQuyNam'
os.makedirs(directory, exist_ok=True)  # Create the directory if it doesn't exist

# Tạo danh sách URL từ các giá trị biến
luuchuyentructiep = "luu-chuyen-tien-te-truc-tiep-chinh-"
baocao = "bao-cao-tai-chinh-"
hoatdongkinhdoanh = "ket-qua-hoat-dong-kinh-doanh-"
luuchuyengiantiep = "luu-chuyen-tien-te-gian-tiep-"
quy = "_4quy"
nam = "_4nam"
luyke = "_luyke6thang"
# Đọc dữ liệu từ tệp JSON
with open('/content/drive/MyDrive/Bigdata_Project/Craw_KeHoachKinhDoanh/URL_request_To_craw_report_Finanal/jsonDataInfor.json') as json_file:
    json_data = json.load(json_file)
# with open('/content/jsonConlai.json') as json_file:
#     json_data = json.load(json_file)
for item in json_data:
    symbol = item["Symbol"]
    uname = item.get("URL", "").split("/")[-1].split(".chn")[0]
    # Loại bỏ giá trị của symbol trong uname
    uname = uname.replace(symbol + "-", "")

    baocaotaichinh = baocao + uname + ".chn"
    luuchuyentienTrucTiep = luuchuyentructiep + uname + ".chn"
    ketquaKinhDoanh = hoatdongkinhdoanh + uname + ".chn"
    luuchuyentienGianTiep = luuchuyengiantiep + uname + ".chn"
    urls = []
    url1 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/BSheet/2023/0/0/0/0/{baocaotaichinh}"
    urls.append(url1)

    url2 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/BSheet/2023/2/0/0/3/{baocaotaichinh}"
    urls.append(url2)

    url3 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/IncSta/2023/2/0/0/{baocaotaichinh}"
    urls.append(url3)

    url4 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/IncSta/2023/0/0/0/0/{ketquaKinhDoanh}"
    urls.append(url4)

    url5 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/IncSta/2023/2/0/0/3/{ketquaKinhDoanh}"
    urls.append(url5)

    url6 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/CashFlow/2023/2/0/0/{ketquaKinhDoanh}"
    urls.append(url6)

    url7 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/CashFlow/2023/0/0/0/0/{luuchuyentienGianTiep}"
    urls.append(url7)

    url8 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/CashFlow/2023/2/0/0/3/{luuchuyentienGianTiep}"
    urls.append(url8)

    url9 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/CashFlowDirect/2023/2/0/0/{luuchuyentienGianTiep}"
    urls.append(url9)
    url10 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/BSheet/2023/0/0/0/0{luuchuyentienTrucTiep}"
    urls.append(url10)
    url11 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/CashFlowDirect/2023/2/0/0/3/{luuchuyentienTrucTiep}"
    urls.append(url11)

    url12 = f"https://s.cafef.vn/bao-cao-tai-chinh/{symbol}/CashFlowDirect/2023/2/0/0/{luuchuyentienTrucTiep}"
    urls.append(url12)

    count = 0
    for url in urls:
        count += 1
        try:
            x = requests.get(url)
            htmldiv = BeautifulSoup(x.content)

            tableData = htmldiv.find("table", {"id": "tableContent"})

            cellHeaders = htmldiv.find("table", {"id": "tblGridData"}).find("tr").find_all("td")[1:5]
            jsonTitle = []

            if count  == 1 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,baocaotaichinh,nam))
            if count  == 2 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,baocaotaichinh,luyke))
            if count  == 3 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,baocaotaichinh,quy))
            if count  == 4 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,ketquaKinhDoanh,nam))
            if count  == 5 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,ketquaKinhDoanh,luyke))
            if count  == 6 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,ketquaKinhDoanh,quy))
            if count  == 7 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,luuchuyengiantiep,nam))
            if count  == 8 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,luuchuyengiantiep,luyke))
            if count  == 9 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,luuchuyengiantiep,quy))
            if count  == 10 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,luuchuyentienTrucTiep,nam))
            if count  == 11 :
                jsonTitle.append("{}_Data{}_{}".format(symbol,luuchuyentienTrucTiep,luyke))
            if count  == 12:
                jsonTitle.append("{}_Data{}_{}".format(symbol,luuchuyentienTrucTiep,quy))

            jsonHeader = []
            for cellHeader in cellHeaders:
                jsonHeader.append(cellHeader.text.strip())

            linesTable = tableData.find_all("tr")

            jsonData = []

            for lineTable in linesTable:
                cellTables = lineTable.find_all("td", limit=5)
                jsonData_line = []
                for cellTable in cellTables:
                    # Chuẩn hóa nội dung JSON
                    text = unicodedata.normalize("NFKD", cellTable.text.strip())
                    jsonData_line.append(text)
                jsonData.append(jsonData_line)

            jsonData.insert(0, [*jsonTitle])
            jsonData.insert(1, ["", *jsonHeader])

            folder_path = "/content/drive/MyDrive/Bigdata_Project/TongHopBaoCaoDoanhNghiepQuyNam/{}_tonghopbaocaodoanhnghiep".format(symbol)
            os.makedirs(folder_path, exist_ok=True)

            file_path = "{}/BaoCaoTaiChinh_{}.json".format(folder_path, jsonTitle)

            with open(file_path, "w", encoding="utf-8") as f:
                jsonDataFormatted = [row for row in jsonData if any(row)]
                json.dump(jsonDataFormatted, f, ensure_ascii=False, indent=4)

            print(f"File saved successfully: {jsonTitle}")
        except Exception as e:
            print(f"Error occurred while processing {url}: {e}")

print("Completed crawl")
```
