```python
import requests
from bs4 import BeautifulSoup
import json

url = "https://s.cafef.vn/hose/DIG-tong-cong-ty-co-phan-dau-tu-phat-trien-xay-dung.chn"
# Tải nội dung của trang web
response = requests.get(url)
html_content = response.text

# Phân tích nội dung HTML
soup = BeautifulSoup(html_content, "html.parser")

# Tạo một dictionary để lưu các thông tin
result = {}

# Tìm thẻ div có class "dltl-wrap"
div = soup.find("div", class_="dltl-wrap")
if div is not None:
    try:
        # Lưu các thông tin vào dictionary
        result["Cập nhật"] = div.find("div", class_="dltlu-time").text.strip().replace("Cập nhật:", "")
    except AttributeError:
        result["Cập nhật"] = "N/A"

    # Cập nhật thông tin từ thẻ div "dl-title clearfix"
    div_dl_title = soup.find("div", class_="dl-title clearfix")
    if div_dl_title is not None:
        symbol_box = div_dl_title.find("div", id="symbolbox")
        if symbol_box is not None:
            result["Mã công ty"] = symbol_box.text.strip()

        name_box = div_dl_title.find("div", id="namebox")
        if name_box is not None:
            company_name = name_box.find("h1").text.strip()
            result["Tên công ty"] = company_name

    try:
        result["Giá điểm"] = div.find("div", class_="dltlu-point").text.strip()
    except AttributeError:
        result["Giá điểm"] = "N/A"

    try:
        thay_doi_div = div.find("div", class_=["dltlu-up green", "dltlu-down red"])
        if thay_doi_div is not None:
            result["Thay đổi"] = thay_doi_div.text.strip()
        else:
            result["Thay đổi"] = "N/A"
    except AttributeError:
        result["Thay đổi"] = "N/A"

    try:
        result["Khối lượng"] = div.find("div", class_="v2").text.strip()
    except AttributeError:
        result["Khối lượng"] = "N/A"

    try:
        result["Giá tham chiếu"] = div.find("div", id="REF").text.strip()
    except AttributeError:
        result["Giá tham chiếu"] = "N/A"

    try:
        result["Giá trần"] = div.find("div", id="CE").text.strip()
    except AttributeError:
        result["Giá trần"] = "N/A"

    try:
        result["Giá sàn"] = div.find("div", id="FL").text.strip()
    except AttributeError:
        result["Giá sàn"] = "N/A"

    try:
        result["Giá mở cửa"] = div.find("div", class_="right up").text.strip()
    except AttributeError:
        result["Giá mở cửa"] = "N/A"

    try:
        result["Giá cao nhất"] = div.find_all("div", class_="right up")[1].text.strip()
    except (AttributeError, IndexError):
        result["Giá cao nhất"] = "N/A"

    try:
        result["Giá thấp nhất"] = div.find("div", class_="right down").text.strip()
    except AttributeError:
        result["Giá thấp nhất"] = "N/A"

    try:
        result["Giao dịch NĐTNN"] = div.find("p").text.strip()
    except AttributeError:
        result["Giao dịch NĐTNN"] = "N/A"

    try:
        result["KLGD ròng"] = div.find_all("div", class_="right")[1].text.strip()
    except (AttributeError, IndexError):
        result["KLGD ròng"] = "N/A"

    try:
        result["GT Mua"] = div.find("div", id="FrBuyVal").text.strip()
    except AttributeError:
        result["GT Mua"] = "N/A"

    try:
        result["GT Bán"] = div.find("div", id="FrSellVal").text.strip()
    except AttributeError:
        result["GT Bán"] = "N/A"

    room_con_lai_div = div.find("div", id="ContentPlaceHolder1_ucTradeInfoV3_divRoomNNConlai")
    if room_con_lai_div is not None:
        result["Room còn lại"] = room_con_lai_div.text.strip()
    else:
        result["Room còn lại"] = "N/A"

# Tìm tất cả các thẻ div có class "kehoachkd"
divs = soup.find_all("li", class_="clearfix")

# Lưu các kết quả vào dictionary
for div in divs:
    key_element = div.find("div", class_="l")
    value_element = div.find("div", class_="r")

    if key_element is not None and value_element is not None:
        key = key_element.text.strip()
        value = value_element.text.strip()
        result[key] = value

# Tìm thẻ div có class "companyIntro" để lấy thông tin giới thiệu công ty
div_company_intro = soup.find("div", class_="companyIntro")
if div_company_intro is not None:
    company_intro = div_company_intro.text.strip()
    result["Giới thiệu công ty"] = company_intro

# Lưu kết quả vào file JSON
with open("/content/drive/MyDrive/Bigdata_Project/Data_craw/resultDigFinalFianl.json", "w", encoding="utf-8") as f:
    json.dump(result, f, ensure_ascii=False, indent=4)

print("Kết quả đã được lưu vào file resultDigFinal.json")
```

```python
import json
import os

# Đường dẫn tới file JSON
json_file = "/content/drive/MyDrive/Bigdata_Project/Data_craw/jsonDataListSympoy.json"

# Đọc nội dung từ file JSON
with open(json_file, "r") as file:
    json_data = json.load(file)

# Tạo thư mục để lưu các file JSON theo nhóm
output_directory = "/content/drive/MyDrive/Bigdata_Project/Data_craw/grouped_data_center"
os.makedirs(output_directory, exist_ok=True)

# Tạo một từ điển rỗng để lưu trữ các nhóm
grouped_data = {}

# Duyệt qua từng đối tượng trong danh sách
for obj in json_data:
    center_name = obj["CenterName"]

    # Kiểm tra xem nhóm đã tồn tại trong từ điển hay chưa
    if center_name in grouped_data:
        grouped_data[center_name].append(obj)
    else:
        grouped_data[center_name] = [obj]

# Lưu các nhóm thành các file JSON riêng
for center_name, group in grouped_data.items():
    output_file = os.path.join(output_directory, f"{center_name}.json")

    # Lưu nội dung vào file JSON
    with open(output_file, "w") as file:
        json.dump(group, file, indent=4, ensure_ascii=False)

    print(f"Lưu nhóm {center_name} vào file {output_file}")

print("Hoàn thành lưu dữ liệu thành các file JSON riêng.")
```
